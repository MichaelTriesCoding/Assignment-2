```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
!pip install pygam
from pygam import LinearGAM, s, f
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

# Read the CSV file
df = pd.read_csv('/content/drive/MyDrive/Baseball Data/NCAA_D1_Trackman_2022_Modified.csv')

# Filter out N/A values for necessary columns
filtered_df = df.dropna(subset=['Angle', 'ExitSpeed', 'PlayResult', 'Distance'])

# Filter the DataFrame based on the PlayResult values
filtered_df = filtered_df[filtered_df['PlayResult'].isin(['Out', 'Single', 'Double', 'Triple', 'HomeRun', 'Error'])]

# Replace "N/A" with 0 in the 'Distance' column
filtered_df['Distance'] = filtered_df['Distance'].fillna(0)

# Step 1: Define the wOBA mapping based on PlayResult
woba_values = {
    'Single': 0.88,
    'Double': 1.25,
    'Triple': 1.57,
    'HomeRun': 2.01,
    'Out': 0.00,
}

# Step 2: Create a new column 'wOBA' in the DataFrame based on the PlayResult
filtered_df['wOBA'] = filtered_df['PlayResult'].map(woba_values).fillna(0)

# Split data into features and target, including 'Distance'
x = filtered_df[['Angle', 'ExitSpeed', 'Distance']]
y = filtered_df['wOBA']

# Split data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Import Linear Regression model & train
model = LinearRegression()
model.fit(x_train, y_train)

# Create and train the GAM model
gam = LinearGAM(s(0) + s(1) + s(2)).fit(x_train, y_train)  # Include 'Distance' in GAM

# Create and train the Random Forest model
rf_model = RandomForestRegressor(n_estimators=10, random_state=42)
rf_model.fit(x_train, y_train)

# Create new data for prediction, including 'Distance'
PlayerA = pd.DataFrame({'Angle': [8], 'ExitSpeed': [90.9], 'Distance': [200]})
PlayerB = pd.DataFrame({'Angle': [9.9], 'ExitSpeed': [81.5], 'Distance': [150]})
Team = pd.DataFrame({'Angle': [13.9], 'ExitSpeed': [86.7], 'Distance': [200]})

# Make predictions using all models
PlayerA_OutcomeLR = model.predict(PlayerA)
PlayerA_OutcomeGAM = gam.predict(PlayerA)
PlayerA_OutcomeRF = rf_model.predict(PlayerA)

PlayerB_OutcomeLR = model.predict(PlayerB)
PlayerB_OutcomeGAM = gam.predict(PlayerB)
PlayerB_OutcomeRF = rf_model.predict(PlayerB)

Team_OutcomeLR = model.predict(Team)
Team_OutcomeGAM = gam.predict(Team)
Team_OutcomeRF = rf_model.predict(Team)

print("Player A Linear Regression wOBA: ", PlayerA_OutcomeLR)
print("Player B Linear Regression wOBA: ", PlayerB_OutcomeLR)
print("Team     Linear Regression wOBA: ", Team_OutcomeLR)
print("Player A GAM wOBA: ", PlayerA_OutcomeGAM)
print("Player B GAM wOBA: ", PlayerB_OutcomeGAM)
print("Team     GAM wOBA: ", Team_OutcomeGAM)
print("Player A Random Forest wOBA: ", PlayerA_OutcomeRF) # Print Random Forest predictions
print("Player B Random Forest wOBA: ", PlayerB_OutcomeRF) # Print Random Forest predictions
print("Team     Random Forest wOBA: ", Team_OutcomeRF) # Print Random Forest predictions
```
