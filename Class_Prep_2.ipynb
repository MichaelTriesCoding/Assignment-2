```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.patches import Rectangle


df = pd.read_csv("/content/drive/MyDrive/Baseball/CSV_Files/NCAA_D1_Trackman_2022_Modified.csv") # try a different encoding like latin-1

df.head()
df.info()
```

```python
for column in df.columns:
  print(f"Unique values in {column}: {df[column].unique()}")
```

```python
fig,ax = plt.subplots()
sns.scatterplot(x = df ['Angle'], y = df ['ExitSpeed'])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x = df ['Angle'], y = df ['ExitSpeed'], hue = df ['PlayResult'])
plt.title('Angle vs. ExitSpeed (Colored by PlayResults)')
plt.show()
```
